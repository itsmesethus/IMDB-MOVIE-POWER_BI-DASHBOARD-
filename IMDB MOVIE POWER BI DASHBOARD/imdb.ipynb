```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/01/Untitled folder/IMDB-Movie-Data.csv")
df
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.drop(columns=['Description'],inplace=True)
df.columns
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(),annot=True,cmap='Blues')
plt.show()
```

```python
df.isnull().sum()
```

```python
df['Revenue (Millions)'].describe(percentiles=[0.35,0.45])
```

```python
per=np.nanpercentile(df['Revenue (Millions)'],30)
df['Revenue (Millions)'].fillna(per,inplace=True)
```

```python
df['Revenue (Millions)'].isna().sum()
```

```python
df['Metascore'].describe()
```

```python
p45=np.nanpercentile(df['Metascore'],45)
df['Metascore'].fillna(p45,inplace=True)
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(),annot=True,cmap='Blues')
plt.show()
```
